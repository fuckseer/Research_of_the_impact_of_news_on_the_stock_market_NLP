```python
from gensim.models import Word2Vec
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report,accuracy_score

# Обучение модели Word2Vec
word2vec_model = Word2Vec(sentences=df['Text_preprocessed'], vector_size=100, window=5, min_count=1, workers=4)

# Преобразование каждого текста в вектор Word2Vec
def text_to_vector(text):
    # Разбиваем текст на токены
    tokens = text.split()
    # Инициализируем счетчик
    count = 0
    # Инициализируем вектор
    vector = np.zeros(100)
    # Проходим по всем токенам в тексте
    for token in tokens:
        # Если токен есть в модели Word2Vec, добавляем его вектор к общему вектору
        if token in word2vec_model.wv:
            vector += word2vec_model.wv[token]
            count += 1
    # Усредняем вектор по всем токенам
    if count != 0:
        vector /= count
    return vector

# Преобразуем каждый текст в вектор
X = np.array([text_to_vector(text) for text in df['Text_preprocessed']])

# Разделим данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, df['Target'], test_size=0.2, random_state=42)

# Обучим модель CatBoost с использованием полученных векторов
catboost_params = {
    'iterations': 300,
    'learning_rate': 0.01,
    'eval_metric': 'Accuracy',
    'early_stopping_rounds': 20,
    'use_best_model': True,
    'verbose': 50
}
model = CatBoostClassifier(**catboost_params)
model.fit(X_train, y_train, eval_set=(X_test, y_test))

# Оценим качество модели
pred = model.predict(X_test)
print(classification_report(y_test, pred))
print("Accuracy:", accuracy_score(y_test, pred))
```
