```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('final_data.csv')
```

```python
wordcloud_params = {
    'width': 800,
    'height': 400,
    'background_color': 'white',
    'collocations': False,
    'random_state': 42
}
```

```python
# Получение общей частоты токенов для всех категорий
all_keywords = df['Tokens'].apply(lambda x: ' '.join(x))
#all_tokens = all_keywords.apply(lambda x: x.split())
total_token_counts = Counter(all_tokens)

# Определение порогового значения для общих токенов
threshold = 0.10 * 12000  # Например, если более 50% категорий содержат токен, он считается общим

# Создание и отображение облака слов для каждой категории target
for target in df['Target'].unique():
    if target == 1:
        # Фильтрация данных по категории target
        target_keywords = df.loc[(df['Target'] == target) & (df['Difference'] > 5)]['Tokens']
    elif target == -1:
        target_keywords = df.loc[(df['Target'] == target) & (df['Difference'] < -3)]['Tokens']
    else:
        target_keywords = df.loc[(df['Target'] == target)]['Tokens']
    # Сбор всех токенов для данной категории
    target_tokens = sum(target_keywords.tolist(), []) 
    
    # Фильтрация токенов, исключая общие токены
    filtered_tokens = [token for token in target_tokens if total_token_counts[token] < threshold]
    
    # Создание объекта WordCloud
    wordcloud = WordCloud(**wordcloud_params).generate(' '.join(filtered_tokens))
    
    # Вывод облака слов
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud)
    plt.title(f'Word Cloud for Target {target}')
    plt.axis('off')  # Отключение осей координат
    plt.show()
```

```python
def clean_keywords(keywords_str):
    # Преобразование строки ключевых слов в список слов
    keywords_list = eval(keywords_str)  # Преобразуем строку в список с помощью eval
    
    # Фильтрация списка ключевых слов: оставляем только непустые слова
    cleaned_list = [word.strip() for word in keywords_list if word.strip()]
    
    return cleaned_list

# Применение функции clean_keywords к столбцу 'Keywords' и создание нового столбца 'Cleaned_Keywords'
df['Tokens'] = df['Tokens'].apply(clean_keywords)
```
